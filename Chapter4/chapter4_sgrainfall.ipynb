```python
%matplotlib inline

import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime
import numpy as np
import pandas as pd

# Load data
data = pd.read_csv("./data/sgrainfall.csv")

# Convert string into Python date time object
data["month"] = pd.to_datetime(data["month"])

date = data["month"]
value = data["total_rainfall"]

year_month_formatter = mdates.DateFormatter("%Y")
month_year_locator = mdates.MonthLocator(interval=1)


fig, ax = plt.subplots(figsize=(8, 6))

ax.plot(data.month, data.total_rainfall, label='Total rainfall', linewidth=1)
ax.xaxis.set_major_formatter(year_month_formatter)
ax.xaxis.set_minor_locator(month_year_locator)

# Add title and axis labels
plt.title('Rainfall in Singapore between Jan 1982 and Jul 2020')
plt.xlabel('Year/Month')
plt.ylabel('Total rainfall (millimetres)')

# Add legend
plt.legend()

fig.autofmt_xdate()
```
